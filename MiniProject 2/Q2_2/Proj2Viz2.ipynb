```python
purpose = aiddata.groupby(["coalesced_purpose_name","coalesced_purpose_code"])["commitment_amount_usd_constant"].sum()
df = pd.DataFrame(purpose)
```

```python
df = df.sort_values(['commitment_amount_usd_constant'], ascending=False)
df.head(11)
```

```python
topten=['Air transport','Rail transport','Industrial development','Power generation/non-renewable sources','RESCHEDULING AND REFINANCING','Import support (capital goods)','Social/ welfare services','Telecommunications','Power generation/renewable sources','Mineral/Metal prospection and exploration']
p = aiddata.groupby(["year","coalesced_purpose_name","coalesced_purpose_code"],as_index=False)["commitment_amount_usd_constant"].sum()
p1=pd.DataFrame(p)
p1.columns=['year','Purpose','Code','amount']
p1.head()
```

```python
p1 = p[p.Purpose.isin(topten)]
p1=p1.sort_values(by=['year'])
p1
```

```python
p1.to_csv("test2data.csv")
```

```python
d2 = p1.groupby(['year', 'Purpose'])['amount'].sum().unstack('Purpose')
```
