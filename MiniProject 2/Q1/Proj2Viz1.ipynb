```python
p = aiddata.groupby(["year","donor"],as_index=False)["commitment_amount_usd_constant"].sum()
p=p.rename(columns={'year':'year','donor':'Country','commitment_amount_usd_constant' : 'donated'})
print(p)
```

```python
p1 = aiddata.groupby(["year","recipient"],as_index=False)["commitment_amount_usd_constant"].sum()
p1=p1.rename(columns={'year':'year','recipient':'Country','commitment_amount_usd_constant' : 'received'})
print(p1)
```

```python
df  = pd.merge(left=p1, right=p, on=['year','Country'], how='outer')
df=df.fillna(0)
```

```python
df.sort_values(by=['Country','year'], inplace=True)
df['total']=df['received']+df['donated']
df.head()
```

```python
df.sort_values(by=['received','year'], inplace=True, ascending=False)
df.head()
```

```python
df.to_csv("P2Viz1.csv")
```
