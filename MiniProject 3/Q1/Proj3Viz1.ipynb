```python
donors = aiddata.groupby(["donor"],as_index=False)['commitment_amount_usd_constant'].sum()
donors.columns=['donor','amount']
```

```python
donors=donors.sort_values(by = ["amount"], ascending=False)
donors_f=donors.head(20)
donors_f
```

```python
donorlist = donors_f['donor'].tolist()
donorlist
```

```python
recipients = aiddata.groupby(["recipient"],as_index=False)['commitment_amount_usd_constant'].sum()
recipients.columns=['recipient','amount']
recipients=recipients.sort_values(by = ["amount"], ascending=False)
recep_f=recipients.head(10)
recep_f
```

```python
receplist = recep_f['recipient'].tolist()
receplist
```

```python
donors_f.to_csv("donors.csv")
recep_f.to_csv("recipients.csv")
```

```python
data=aiddata.groupby(['donor','recipient'],as_index=False)['commitment_amount_usd_constant'].sum()
data.columns=['donor','recipient','amount']
data.head()
```

```python
data2 = aiddata.groupby(['year','donor','recipient'],as_index=False)['commitment_amount_usd_constant'].sum()
data2.columns=['year','donor','recipient','amount']
data2.head()
```

```python
data_f2 = data2[data2['donor'].isin(donorlist)]
data_f2 = data_f2[data_f2['recipient'].isin(receplist)]
data_f2.head()
```

```python
test = data_f2.groupby('year')
sort_function = lambda x: x.sort_values("amount", ascending = False)
desired_df = test.apply(sort_function)
desired_df.set_index('year')
```

```python
desired_df.to_csv("viz3data.csv")
```
