```python
column_values = (data['donor'].append(data['recipient'])).unique()
#countries =  pd.unique(column_values)
#countries.dtypes
countries=sorted(column_values.tolist())
print(countries)
```

```python
donors=data.groupby(["donor"],as_index=False)["commitment_amount_usd_constant"].sum()
donors=donors.rename(columns={'donor':'Country','commitment_amount_usd_constant' : 'Amount'})
donors['Amount']=.000001*donors['Amount']
Fdonors = donors.assign(Type="donated")
#print(Fdonors)
```

```python
recipients=data.groupby(["recipient"],as_index=False)["commitment_amount_usd_constant"].sum()
recipients=recipients.rename(columns={'donor':'Country','commitment_amount_usd_constant' : 'Amount'})
recipients['Amount']=-.000001*recipients['Amount']
Frecipients = recipients.assign(Type="received")
#print(len(Frecipients))
```

```python
fd=sorted(Fdonors.values.tolist()) 
fr=sorted(Frecipients.values.tolist())
```

```python
def convertToList(a):
    l=[]
    for i in range(len(a)):
        l.append(a[i][0])
    return l
```

```python
a = set(countries).symmetric_difference(set(convertToList(fd)))
print(list(a))
```

```python
for i in a:
    fd.append([i,0,"donated"])
```

```python
print(fd)
```

```python
b = set(countries).symmetric_difference(set(convertToList(fr)))
print(list(b))
```

```python
for i in b:
    fr.append([i,0,"received"])
```

```python
print(fr)
```

```python
Donors = pd.DataFrame(fd,columns=['Country',"Amount","Type"])
Recipients = pd.DataFrame(fr,columns=["Country","Amount","Type"])
finaldata = pd.concat([Donors, Recipients], axis=0)
finaldata.sort_values(by=['Country'], inplace=True)
```

```python
finaldata = finaldata.reset_index()
del finaldata['index']
```

```python
finaldata.to_csv('Viz1data.csv')
```
